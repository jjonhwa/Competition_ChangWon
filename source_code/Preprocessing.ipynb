### 인구수 데이터 정리

```python
pop = pd.read_excel('../data/인구.xlsx', header = None)
pop.head()
```

```python
pop = pop.iloc[:, :2]
pop.head()
```

```python
pop.columns = ['읍면동', '총 인구수']
pop.head()
```

```python
final_name = []
final_pop = []

for i in range(len(pop['읍면동'])) :
    tmp_name = pop['읍면동'][i].replace(' ', '')
    tmp_pop = pop['총 인구수'][i]    
    
    if tmp_name[-1] == '읍' :
        final_name.append(tmp_name)
        final_pop.append(tmp_pop)
    if tmp_name[-1] == '면' :
        final_name.append(tmp_name)
        final_pop.append(tmp_pop)
    if tmp_name[-1] == '동' :
        final_name.append(tmp_name)
        final_pop.append(tmp_pop)
```

```python
population = pd.DataFrame(final_name, columns = ['읍면동'])
population['총인구수'] = final_pop
population.head()
```

```python
population.index = population['읍면동']
population.drop(['읍면동'], axis = 1, inplace = True)
population.head()
```

### 공장데이터

```python
fac = pd.read_csv('../data/공장.csv', encoding = 'cp949')
fac
```

```python
import googlemaps
from tqdm import tqdm_notebook
```

```python
gmaps_key = 'AIzaSyAAdqWF_Ykxo8Fcn7FJmrSLfG3LJojryhQ'
gmaps = googlemaps.Client(key = gmaps_key)
```

```python
fac_address = fac['주소']
```

```python
fac_dong = []

for i in tqdm_notebook(range(len(fac_address))) :
    tmp = gmaps.geocode(fac_address[i], language = 'ko')
    try :
        tmp_tmp = tmp[0].get('formatted_address').split(' ')
        
        for j in range(len(tmp_tmp)) :
            if tmp_tmp[j][-1] == '면' :
                fac_dong.append(tmp_tmp[j])
                break
            if tmp_tmp[j][-1] == '읍' :
                fac_dong.append(tmp_tmp[j])
                break
            if tmp_tmp[j][-1] == '동' :
                fac_dong.append(tmp_tmp[j])
                break
                
    except :
        print(fac_address[i])
```

```python
factory = pd.DataFrame(fac_dong, columns = ['읍면동'])
factory['공장수'] = 1
```

```python
factory = pd.pivot_table(factory, index = ['읍면동'], aggfunc = 'sum')
factory.head()
```

### 공영주차장

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = 'https://www.changwon.go.kr/depart/contents.do?mId=0707020100'

page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
```

```python
table = soup.select('div.conts')[0].find_all('div')
```

* 우리가 필요한 table은 1~5번째 테이블(각 구별 공영주차장 수) 

```python
parking_lot = []

for i in range(1,6) :
    tmp = table[i].find_all('td')
    
    for j in range(len(tmp)) :
        if tmp[j].string != None :
            tmp_tmp = tmp[j].string
        
            if tmp_tmp[-1] == '읍' :
                parking_lot.append(tmp_tmp)
            if tmp_tmp[-1] == '면' :
                parking_lot.append(tmp_tmp)
            if tmp_tmp[-1] == '동' : 
                parking_lot.append(tmp_tmp)
```

```python
len(parking_lot)
```

```python
parking = pd.DataFrame(parking_lot, columns = ['읍면동'])
parking['공영주차장'] = 1
```

```python
parking = pd.pivot_table(parking, index = ['읍면동'], aggfunc = 'sum')
parking.head()
```
